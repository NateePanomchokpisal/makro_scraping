```python
from selenium import webdriver
from time import sleep
import pandas as pd

URL = 'https://www.makroclick.com/th/category/vegetables-and-fruits'
browser = webdriver.Firefox(executable_path = r'C:\Users\DELL\geckodriver.exe')
browser.get(URL)
sleep(1)

ending_page = 24
page = 1
while page <= ending_page:
    main = browser.find_element_by_css_selector('div.pt-2.col-12')
    products = main.find_elements_by_css_selector('div.ProductCard__CardProduct-sc-1lb6pe3-0.ktTHor')
    product_list = []
    for product in products:
        product_id = product.find_element_by_css_selector("div.ProductCard__TextCode-sc-1lb6pe3-3.kboYsK.my-1")
        print(product_id.text)
        product_name = product.find_element_by_css_selector("div.ProductCard__TitleMaxLines-sc-1lb6pe3-2.kAKAhd.pt-2")
        print(product_name.text)
        product_SKUprice =product.find_element_by_css_selector("div.ProductCard__TextPrice-sc-1lb6pe3-1.jcvzXx.m-0.mt-lg-1.font-16.color-red.weight-bold")
        print(product_SKUprice.text)
        product_PIECEprice =product.find_element_by_css_selector("div.font-12.color-darkgrey.d-flex.flex-row")
        print(product_PIECEprice.text)
        images = product.find_elements_by_tag_name("img")
        for image in images:
            img_list = []
            if image.get_attribute('src') and 'https' in image.get_attribute('src'):
                print(image.get_attribute('src'))
                img_list.append(image.get_attribute('src'))
            product_list.append([product_id.text,product_name.text,product_SKUprice.text,product_PIECEprice.text,img_list])
    
    try:
        next_page = browser.find_element_by_xpath('/html/body/div[1]/div/div/div[3]/div/div/div[1]/div[2]/div[2]/div[1]/div[3]/div/div[2]/div[2]/div/div[6]')
        next_page.click()
    except:
        next_page = browser.find_element_by_xpath('/html/body/div[1]/div/div/div[3]/div/div/div[1]/div[2]/div[2]/div[1]/div/div[3]/div/div[2]/div/div[7]')
        next_page.click()
    sleep(2)
    page+=1
    
browser.close()
```

```python
df = pd.DataFrame(product_list,columns = ["Id","Name","SKU_price","Piece_price","Image"])
df["Id"] = df["Id"].str[-7:]
df["Piece_price"] = df['Piece_price'].str.replace("ราคาต่อชิ้น","")
df["Piece_price"] = df['Piece_price'].str.replace("฿","")
df['SKU_price'] = df['SKU_price'].str.replace("฿","")
df["Image"] =df["Image"].str[0]
df
```
